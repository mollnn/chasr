```python
import python_speech_features as p
import scipy.io.wavfile as wav
import glob
import numpy as np
from os.path import join
from tqdm import tqdm
from tensorflow.keras.preprocessing.sequence import pad_sequences
import os
import librosa
```

```python
path_base = './dataset/data_thchs30'
path_data = join(path_base, 'data')

wav_files = glob.glob(join(path_data,'*.wav'))
wav_files.sort()
wav_files
```

```python
wav_files = wav_files[:2]   # for demo
wav_files
```

```python
wav_file = wav_files[0]
fs, audio = wav.read(wav_file)
r = p.mfcc(audio, samplerate=fs, numcep=26)
r.shape
```

```python
t = np.transpose(r)
x = pad_sequences(t, maxlen=778, dtype='float', padding='post', truncating='post').T 
x.shape
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(20,5))
plt.imshow(x.T, cmap='jet')
```
